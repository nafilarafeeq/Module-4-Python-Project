```python
bins = [20, 30, 40, 50, 60, 70]
labels = ['20-30', '31-40', '41-50', '51-60', '61-70']
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


age_group_distribution = df['Age_group'].value_counts()

print(age_group_distribution)
```

```python
age_group_distribution.plot(kind='bar', color='lightgreen')
plt.title('Predominant Age Group Among Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
team_salary_expenditure = df.groupby('Team')['Salary'].sum()
position_salary_expenditure = df.groupby('Position')['Salary'].sum()


highest_team_salary = team_salary_expenditure.idxmax(), team_salary_expenditure.max()
highest_position_salary = position_salary_expenditure.idxmax(), position_salary_expenditure.max()

print(f"Team with the highest salary expenditure: {highest_team_salary}")
print(f"Position with the highest salary expenditure: {highest_position_salary}")
```

```python
team_salary_expenditure.plot(kind='bar', color='orange')
plt.title('Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


position_salary_expenditure.plot(kind='bar', color='purple')
plt.title('Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
